```python
import os # ÌôòÍ≤Ω Î≥ÄÏàòÎÇò ÎîîÎ†âÌÑ∞Î¶¨, ÌååÏùº Îì±Ïùò OS ÏûêÏõêÏùÑ Ï†úÏñ¥Ìï† Ïàò ÏûàÍ≤å Ìï¥Ï£ºÎäî Î™®Îìà
import cv2 # OpenCVÎùºÏù¥Î∏åÎü¨Î¶¨ ‚Üí Ïª¥Ìì®ÌÑ∞ ÎπÑÏ†Ñ Í¥ÄÎ†® ÌîÑÎ°úÍ∑∏ÎûòÎ∞çÏùÑ ÏâΩÍ≤å Ìï† Ïàò ÏûàÎèÑÎ°ù ÎèÑÏôÄÏ£ºÎäî ÎùºÏù¥Î∏åÎü¨Î¶¨
import matplotlib.pyplot as plt # Îã§ÏñëÌïú Îç∞Ïù¥ÌÑ∞Î•º ÎßéÏùÄ Î∞©Î≤ïÏúºÎ°ú ÎèÑÏãùÌôî Ìï† Ïàò ÏûàÎèÑÎ°ù ÌïòÎäî ÎùºÏù¥Î∏åÎü¨Î¶¨
import numpy as np # Îã§Ï∞®Ïõê Î∞∞Ïó¥ÏùÑ ÏâΩÍ≤å Ï≤òÎ¶¨ÌïòÍ≥† Ìö®Ïú®Ï†ÅÏúºÎ°ú ÏÇ¨Ïö©Ìï† Ïàò ÏûàÎèÑÎ°ù ÌïòÎäî ÎùºÏù¥Î∏åÎü¨Î¶¨
import dlib # Ïù¥ÎØ∏ÏßÄ Ï≤òÎ¶¨ Î∞è Í∏∞Í≥Ñ ÌïôÏäµ, ÏñºÍµ¥Ïù∏Ïãù Îì±ÏùÑ Ìï† Ïàò ÏûàÎäî c++ Î°ú Í∞úÎ∞úÎêú Í≥†ÏÑ±Îä•Ïùò ÎùºÏù¥Î∏åÎü¨Î¶¨
print("üå´üõ∏")
```

```python
selfie_path = 'camera_sticker/images/Selfie.png'
selfie_bgr = cv2.imread(selfie_path)
selfie_rgb = cv2.cvtColor(selfie_bgr, cv2.COLOR_BGR2RGB)
selfie_show = selfie_bgr.copy()
plt.imshow(selfie_rgb)
plt.show()
```

```python
# dlibÏùò Í∏∞Î≥∏ ÏñºÍµ¥ Í∞êÏßÄÍ∏∞ instance ÏÉùÏÑ±
detector_hog = dlib.get_frontal_face_detector()
dlib_rects = detector_hog(selfie_rgb, 1)
```

```python
print(dlib_rects)
```

```python
# selfie_show Ïù¥ÎØ∏ÏßÄÏóê bounding box Ï†ÅÏö©
for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    r = dlib_rect.right()
    t = dlib_rect.top()
    b = dlib_rect.bottom()

    cv2.rectangle(selfie_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)

selfie_show_rgb = cv2.cvtColor(selfie_show, cv2.COLOR_BGR2RGB)
plt.imshow(selfie_show_rgb)
plt.show()
```

```python
# Ï†ÄÏû•Ìïú landmark Î™®Îç∏ Î∂àÎü¨Ïò§Í∏∞
model_path = '/aiffel/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'

landmark_predictor = dlib.shape_predictor(model_path)
```

```python
# ÏÇ¨ÏßÑÏóê ÏûàÎäî Í∞Å ÏñºÍµ¥Ïóê Ï∞çÏùÄ Ï†êÎì§Ïùò Ï¢åÌëúÎ•º Ï†ÄÏû•ÌïòÎäî Î¶¨Ïä§Ìä∏
# Ìïú Í∞úÏùò ÏñºÍµ¥ Îãπ ÌïòÎÇòÏùò Î¶¨Ïä§Ìä∏Ïóê Ï†ê Ï¢åÌëú Ï†ÄÏû•
# 3 ÏÇ¨ÎûåÏù¥ ÎÇòÏôîÏúºÎ©¥ list_landmarksÎäî 3Í∞úÏùò Î¶¨Ïä§Ìä∏Î•º Îã¥Îäî Î¶¨Ïä§Ìä∏.
list_landmarks = []

for dlib_rect in dlib_rects:
    points = landmark_predictor(selfie_rgb, dlib_rect)
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)

# ÏñºÍµ¥Ïóê Î™á Í∞úÏùò Ï†ê Ï∞çÏóàÎäîÏßÄ ÌôïÏù∏
print(len(list_landmarks[0]))
```

```python
# Ï¢åÌëú Î¶¨Ïä§Ìä∏Îì§ÏùÑ Îã¥ÏùÄ Î¶¨Ïä§Ìä∏ÏóêÏÑú Ï¢åÌëú Î¶¨Ïä§Ìä∏Î•º ÎΩëÏïÑÏÑú
for landmark in list_landmarks:
    # Ï¢åÌëú Î¶¨Ïä§Ìä∏Ïùò Ï¢åÌëúÎßàÎã§
    for point in landmark:
        # cv2.circle() methodÎ°ú ÏõêÏùÑ Í∑∏Î†§Ï§å.
        cv2.circle(selfie_show, point, 2, (0,255, 255), 6)

# RGB Ï†ÑÌôò
selfie_show_rgb = cv2.cvtColor(selfie_show, cv2.COLOR_BGR2RGB)
plt.imshow(selfie_show_rgb)
plt.show()
```

```python
# Ïä§Ìã∞Ïª§Ïùò Ï§ëÏã¨Ïù¥ Îê† Ï†êÏùò Ï¢åÌëúÏôÄ ÏñºÍµ¥ ÎÑàÎπÑ Í∞í ÌôïÏù∏
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    # ÏΩîÏùò index: 30.
    x = landmark[30][0]
    y = landmark[30][1]
    w = h = dlib_rect.width()
    # ÏΩîÏùò x,y Ï¢åÌëú
    print(f'(x,y): ({x},{y})')
    # ÏñºÍµ¥ ÎÑàÎπÑ
    print(f'(w,h): ({w},{h})')
```

```python
# Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄ Î∂àÎü¨Ïò§Í∏∞
sticker_path = "camera_sticker/images/Cat Whiskers Icon.png"
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w,h))
print(img_sticker.shape)
```

```python
refined_x = x - w // 2
refined_y = y - 667//2
print(refined_x, refined_y)
```

```python
plt.imshow(selfie_rgb)
plt.show()
```

```python
# Ïä§Ìã∞Ïª§Í∞Ä Îì§Ïñ¥Í∞à Î∂ÄÎ∂ÑÏùò ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ Ï†ÄÏû•
sticker_area = selfie_bgr[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
plt.imshow(sticker_area)
plt.show()
```

## Ïä§Ìã∞Ïª§ Ï†ÅÏö©

```python
# ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ selfie_bgrÏóê ÎåÄÌï¥ np.whereÌï®ÏàòÎ•º Ïù¥Ïö©Ìï¥ÏÑú
# Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄÏóêÏÑú Ïä§Ìã∞Ïª§ Ïô∏Ïùò Í≥µÎ∞± ÏòÅÏó≠ÏùÑ
# ÏúÑÏóêÏÑú Ï†ÄÏû•Ìï¥ Îëî sticker_area Ïù¥ÎØ∏ÏßÄÎ°ú ÎåÄÏ≤¥
selfie_bgr[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==0,img_sticker,sticker_area).astype(np.uint8)
```

```python
plt.imshow(cv2.cvtColor(selfie_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
print(selfie_show.shape)
print(img_sticker.shape)
```

```python
# Ïä§Ìã∞Ïª§ Ïó∞ÌïòÍ≤å ÎßåÎì§Í∏∞ (cv2.addWeighted() Ìï®Ïàò Ïù¥Ïö©)

# ÏúÑÏóêÏÑú ÌôïÏù∏Ìïú bounding box(dlib_rect)Ïùò Ï¢åÏÉÅÎã® Ï¢åÌëú (206, 576)
x = 206
y = 576
sticker_h, sticker_w, _ = img_sticker.shape

# ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏóêÏÑú Region of Interest ÏÑ§Ï†ï.
roi = selfie_bgr[y:y+sticker_h, x:x+sticker_w]

# Ïä§Ìã∞Ïª§ Ïó∞ÌïòÍ≤å ÎßåÎì§Í∏∞
blended_roi = cv2.addWeighted(roi, 0.75, img_sticker, 0.25, 0.0)

# ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏóêÏÑú roiÏóê Ìï¥ÎãπÌïòÎäî Î∂ÄÎ∂ÑÏùÑ blended_roiÎ°ú ÎåÄÏ≤¥
selfie_show[y:y+sticker_h, x:x+sticker_w] = blended_roi

selfie_show_rgb = cv2.cvtColor(selfie_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(selfie_show_rgb)
plt.show()
```
